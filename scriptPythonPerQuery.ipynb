```python
lista_colonne = []
file1 = open('categories1.txt', 'r')

lines = file1.readlines()
for line in lines:
    lista_colonne.append(line.strip())

lista_query_no_aggiustamento_no_filtro = []
lista_query_no_aggiustamento_filtro = []
lista_query_aggiustamento_no_filtro = []
lista_query_aggiustamento_filtro = []

for colonna in lista_colonne:
    lista_query_no_aggiustamento_no_filtro.append(f"select avg({colonna}) from market.market;")
    for colonna2 in lista_colonne:
        if (colonna != colonna2):
            lista_query_no_aggiustamento_filtro.append(f"select avg({colonna}) from market.market where {colonna2} > 60 allow filtering;")
            lista_query_aggiustamento_no_filtro.append(f"select avg(({colonna}/(100-{colonna2})) * 100 ) from market.market;")
            lista_query_aggiustamento_filtro.append(f"select avg(({colonna}/(100-{colonna2}))*100) from market.market where {colonna2} > 60 allow filtering;")
```
